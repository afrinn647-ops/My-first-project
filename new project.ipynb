```python
summary_starts =Data_file.describe().transpose()
```

```python
summary_starts['range'] = summary_starts['max'] - summary_starts['min']
```

```python
final_starts = summary_starts[['mean', 'std', 'min', 'max', 'range']]
```

```python
final_starts.columns = ['Mean', 'Standard Deviation', 'Minimum', 'Maximum', 'Range']
```

```python
final_starts.to_csv('summary_statistics.csv')
```
